### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



### a.- Ingresos Totales por Tienda

```python
# Sumamos la columna precio
ingresos_tienda = tienda['Precio'].sum()
ingresos_tienda2 = tienda2['Precio'].sum()
ingresos_tienda3 = tienda3['Precio'].sum()
ingresos_tienda4 = tienda4['Precio'].sum()

# Mostramos el valor calculado de la columna precio
print(f'Los ingresos totales de la tienda 1 son: {ingresos_tienda}')
print(f'Los ingresos totales de la tienda 2 son: {ingresos_tienda2}')
print(f'Los ingresos totales de la tienda 3 son: {ingresos_tienda3}')
print(f'Los ingresos totales de la tienda 4 son: {ingresos_tienda4}')
```

```python
# Lista para recorrer
tiendas = [ingresos_tienda, ingresos_tienda2, ingresos_tienda3, ingresos_tienda4]

# Referencia: Tienda 1
tienda_lider = ingresos_tienda

print(f"Tienda 1: ${ingresos_tienda:,.0f} (Tienda con mayor Ingreso)")

for i, ingreso in enumerate(tiendas[1:], start=2):
    diferencia = ((ingreso - tienda_lider) / tienda_lider) * 100
    print(f"Tienda {i}: ${ingreso:,.0f} ({diferencia:.1f}% contra la Tienda 1)")
```

# 2. Ventas por categoría

### a.- Calcular la cantidad de productos vendidos por categoría en cada tienda.

```python
# Calculamos la cantidad de prodcutos vendidos por categoría
ventas_por_categoria_tienda1 = tienda.groupby(['Categoría del Producto']).size().reset_index(name='cantidad_vendida').sort_values(
                                by='cantidad_vendida',ascending=True).reset_index(drop=True)

ventas_por_categoria_tienda2 = tienda2.groupby(['Categoría del Producto']).size().reset_index(name='cantidad_vendida').sort_values(
                                by='cantidad_vendida',ascending=True).reset_index(drop=True)

ventas_por_categoria_tienda3 = tienda3.groupby(['Categoría del Producto']).size().reset_index(name='cantidad_vendida').sort_values(
                                by='cantidad_vendida',ascending=True).reset_index(drop=True)

ventas_por_categoria_tienda4 = tienda4.groupby(['Categoría del Producto']).size().reset_index(name='cantidad_vendida').sort_values(
                                by='cantidad_vendida',ascending=True).reset_index(drop=True)
```

### b.- Mostramos tabla con la cantidad de productos por categoría

```python
print(f'La Cantidad de Productos vendidos por categoría para la Tienda n°1 son: \n\n {ventas_por_categoria_tienda1}')
```

```python
print(f'La Cantidad de Productos vendidos por categoría para la Tienda n°3 son: \n\n {ventas_por_categoria_tienda3}')
```

```python
print(f'La Cantidad de Productos vendidos por categoría para la Tienda n°4 son: \n\n {ventas_por_categoria_tienda4}')
```

# 3. Calificación promedio de la tienda


### a.- Calificación de los clientes por tienda

```python
# Conocer la Calificación Máxima existente
calificacion_maxima = tienda['Calificación'].max()
print(f'La calificación máxima registrada es: {tienda["Calificación"].max()}')
```

```python
# Calculamos y mostramos los promedio de califiación por cada Tienda
promedio_calificacion_tienda1 = (tienda['Calificación'].mean())
promedio_calificacion_tienda2 = (tienda2['Calificación'].mean())
promedio_calificacion_tienda3 = (tienda3['Calificación'].mean())
promedio_calificacion_tienda4 = (tienda4['Calificación'].mean())

print(f'La Calificación para la Tienda N°1 es: {promedio_calificacion_tienda1:.2f}')
print(f'La Calificación para la Tienda N°2 es: {promedio_calificacion_tienda2:.2f}')
print(f'La Calificación para la Tienda N°3 es: {promedio_calificacion_tienda3:.2f}')
print(f'La Calificación para la Tienda N°4 es: {promedio_calificacion_tienda4:.2f}')
```

# 4. Productos más y menos vendidos

### a.- Producto que mas se venden por tienda

```python
# Producto mas vendido por tienda

# Contamos las veces que aparece cada producto en nuestro DataFrame (En Cada Tienda)
ventas_por_producto_tienda1 = tienda['Producto'].value_counts()
ventas_por_producto_tienda2 = tienda2['Producto'].value_counts()
ventas_por_producto_tienda3 = tienda3['Producto'].value_counts()
ventas_por_producto_tienda4 = tienda4['Producto'].value_counts()

# Buscamos el producto que tiene la cantidad más alta
producto_mas_vendido_tienda1 = ventas_por_producto_tienda1.idxmax()
producto_mas_vendido_tienda2 = ventas_por_producto_tienda2.idxmax()
producto_mas_vendido_tienda3 = ventas_por_producto_tienda3.idxmax()
producto_mas_vendido_tienda4 = ventas_por_producto_tienda4.idxmax()

# Obtienemos el valor numérico del prodcuto mas vendido
cantidad_mas_vendida_tienda1 = ventas_por_producto_tienda1.max()
cantidad_mas_vendida_tienda2 = ventas_por_producto_tienda2.max()
cantidad_mas_vendida_tienda3 = ventas_por_producto_tienda3.max()
cantidad_mas_vendida_tienda4 = ventas_por_producto_tienda4.max()

# Mostramos los resultados obtenidos
print(f'El producto mas vendido en la Tienda n°1 es: {producto_mas_vendido_tienda1}, con {cantidad_mas_vendida_tienda1} unidades')
print(f'El producto mas vendido en la Tienda n°2 es: {producto_mas_vendido_tienda2}, con {cantidad_mas_vendida_tienda2} unidades')
print(f'El producto mas vendido en la Tienda n°3 es: {producto_mas_vendido_tienda3}, con {cantidad_mas_vendida_tienda3} unidades')
print(f'El producto mas vendido en la Tienda n°4 es: {producto_mas_vendido_tienda4}, con {cantidad_mas_vendida_tienda4} unidades')
```

### b.- Producto que menos se venden por tienda

```python
# Producto menos vendido por tienda

# Buscamos el producto que tiene la cantidad más alta
producto_menos_vendido_tienda1 = ventas_por_producto_tienda1.idxmin()
producto_menos_vendido_tienda2 = ventas_por_producto_tienda2.idxmin()
producto_menos_vendido_tienda3 = ventas_por_producto_tienda3.idxmin()
producto_menos_vendido_tienda4 = ventas_por_producto_tienda4.idxmin()

# Obtienemos el valor numérico del prodcuto mas vendido
cantidad_menos_vendida_tienda1 = ventas_por_producto_tienda1.min()
cantidad_menos_vendida_tienda2 = ventas_por_producto_tienda2.min()
cantidad_menos_vendida_tienda3 = ventas_por_producto_tienda3.min()
cantidad_menos_vendida_tienda4 = ventas_por_producto_tienda4.min()

# Mostramos los resultados obtenidos
print(f'El producto menos vendido en la Tienda n°1 es: {producto_menos_vendido_tienda1}, con {cantidad_menos_vendida_tienda1} unidades')
print(f'El producto menos vendido en la Tienda n°2 es: {producto_menos_vendido_tienda2}, con {cantidad_menos_vendida_tienda2} unidades')
print(f'El producto menos vendido en la Tienda n°3 es: {producto_menos_vendido_tienda3}, con {cantidad_menos_vendida_tienda3} unidades')
print(f'El producto menos vendido en la Tienda n°4 es: {producto_menos_vendido_tienda4}, con {cantidad_menos_vendida_tienda4} unidades')
```

# 5. Envío promedio por tienda

### a.- Promedios de envío

```python
# Calculamos el promedio de envío para cada tienda
promedio_envio1 = tienda['Costo de envío'].mean()
promedio_envio2 = tienda2['Costo de envío'].mean()
promedio_envio3 = tienda3['Costo de envío'].mean()
promedio_envio4 = tienda4['Costo de envío'].mean()

#Mostramos el valor calculado
print(f'El valor promedio de envío de la Tienda n°1 es: {promedio_envio1:.2f}')
print(f'El valor promedio de envío de la Tienda n°2 es: {promedio_envio2:.2f}')
print(f'El valor promedio de envío de la Tienda n°3 es: {promedio_envio3:.2f}')
print(f'El valor promedio de envío de la Tienda n°4 es: {promedio_envio4:.2f}')
```

# 6. Visualización de los Datos

### a.- Porcentaje de Ventas por Tienda

```python
import matplotlib.pyplot as plt

# Ingresos por tienda
ingresos = [ingresos_tienda, ingresos_tienda2, ingresos_tienda3, ingresos_tienda4]
labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Organizamos el Gráfico circular
plt.figure(figsize=(7,7))
plt.pie(ingresos,labels=labels,autopct='%1.1f%%',   # Muestra porcentajes con 1 decimal
        startangle=90,                              # Rotación inicial para estética
        wedgeprops={'edgecolor': 'black'})          # Bordes más definidos

# Definimos el título
plt.title("Porcentaje de Ingresos por Tienda")

# Ajustamos espacio de forma automática
plt.tight_layout()
# Llamamos al gráfico
plt.show()
```

### b.- Prodcuto Mas y Menos Vendido

```python
# Productos y cantidades
productos_mas = [producto_mas_vendido_tienda1, producto_mas_vendido_tienda2,
                 producto_mas_vendido_tienda3, producto_mas_vendido_tienda4]

productos_menos = [producto_menos_vendido_tienda1, producto_menos_vendido_tienda2,
                   producto_menos_vendido_tienda3, producto_menos_vendido_tienda4]

cant_mas = [cantidad_mas_vendida_tienda1, cantidad_mas_vendida_tienda2,
            cantidad_mas_vendida_tienda3, cantidad_mas_vendida_tienda4]

cant_menos = [cantidad_menos_vendida_tienda1, cantidad_menos_vendida_tienda2,
              cantidad_menos_vendida_tienda3, cantidad_menos_vendida_tienda4]

labels = ['T1', 'T2', 'T3', 'T4']

# Configuramos posiciones
x = list(range(len(labels)))   # posiciones para cada tienda
width = 0.35  # ancho de cada barra

plt.figure(figsize=(12,6))

# Producto más vendid0 (color verde)
plt.bar([pos - width/2 for pos in x], cant_mas, width, label="Más vendido", color="green")
# Producto menos vendido (color gris)
plt.bar([pos + width/2 for pos in x], cant_menos, width, label="Menos vendido", color="gray")

# Etiquetas en cada barra
for i in range(len(x)):
    plt.text(x[i] - width/2, cant_mas[i] + 0.1, productos_mas[i], ha='center')
    plt.text(x[i] + width/2, cant_menos[i] + 0.1, productos_menos[i], ha='center')

# Definimos la etiquetas para cada eje
plt.xticks(x, labels)
plt.ylabel("Cantidad Vendida")

# Definimos el título
plt.title("Producto Más y Menos Vendido por Tienda")
# Definimos la leyenda
plt.legend()

# Ajustamos espacio de forma automática
plt.tight_layout()
# Llamamos al gráfico
plt.show()
```

### c.-Satisfacción del Cliente

```python
# Datos
tiendas = ['T1', 'T2', 'T3', 'T4']
promedios = [
    promedio_calificacion_tienda1,
    promedio_calificacion_tienda2,
    promedio_calificacion_tienda3,
    promedio_calificacion_tienda4
]

plt.figure(figsize=(8,5))
plt.plot(tiendas, promedios, marker='o', linewidth=2)

plt.title("Satisfacción del Cliente por Tienda")
plt.ylabel("Promedio de Calificación")
plt.ylim(0, 5)
plt.grid(True)

plt.tight_layout()
plt.show()
```

### d.-Costo de Envío por tienda

```python
# Datos
tiendas = ['T1', 'T2', 'T3', 'T4']
promedios_envio = [
    promedio_envio1,
    promedio_envio2,
    promedio_envio3,
    promedio_envio4
]

plt.figure(figsize=(8,5))
plt.fill_between(tiendas, promedios_envio, alpha=0.3)
plt.plot(tiendas, promedios_envio, marker='o')

plt.title("Promedio del Costo de Envío por Tienda")
plt.ylabel("Costo Promedio de Envío")
plt.grid(True)

# Etiquetas por punto
for i in range(len(tiendas)):
    plt.text(tiendas[i], promedios_envio[i] + 0.05,
             f"{promedios_envio[i]:.2f}",
             ha='center')

plt.tight_layout()
plt.show()
```
